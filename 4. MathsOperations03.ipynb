```python
# Aggregation functions // min max mean sum std(standard deviation) var(varians)
arr = np.array([10,20,30,40,50,60,70])
```

```python
print(np.min(arr))
print(np.max(arr))
print(np.mean(arr))
print(np.sum(arr))
print(np.std(arr))
print(np.var(arr))
```

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4,5,6])
print("Array 1:", arr1)
print("Array 2:", arr2)

# Applying a binary operation: add , subtract , multiply, divide, power , mod
add = np.add(arr1,arr2)
sub = np.subtract(arr1,arr2)
div = np.divide(arr1,arr2)
mul = np.multiply(arr1,arr2)
power = np.power(arr1,arr2)
exp = np.exp(arr1)
mod = np.mod(arr1,arr2)
sqrt = np.sqrt(arr2)

print("Addition Result:", add)
print("Subtrtaction Result:", sub)
print("Division Result:", div)
print("Multiplication Result:", mul)
print("Exponentiation Result:", power)
print("Modulus Result:", mod)
print("Exponentent (e to the power) Result:", exp)
print("Square Root Result:", sqrt)
```

```python
a = np.array([[1, 2],[3, 4]])
b = np.array([[4, 3],[2, 1]])
               
print ("Adding 1 to every element:", a + 10)

print ("\nSubtracting 2 from each element:", b - 10)
 
# sum of array elements
# Performing Unary operations
print ("\nSum of all array elements: ",a.sum())
 
# Adding two arrays
# Performing Binary operations
print ("\nArray sum:\n", np.add(a,b))
```

```python
# Matrix multiplication
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
print(np.dot(a,b))
print(a @ b)
```
