```python
import numpy as np
```

```python
# To find the shape of the array.
arr1 = np.array([[1,2,3],[4,5,6]])
print(arr1)
print(arr1.shape)
```

```python
# To find total number of elements in the array.
print(arr1.size)
```

```python
# To find the number of dimention in the array.
print(arr1.ndim)
```

```python
arr_3d = np.array([[[1,4,8],[5,6,2]],[[9,7,8],[1,4,2]]])
print(arr_3d)
print(arr_3d.ndim)
print(arr_3d.shape)
```

```python
# To check the data type of the array.
arrInt = np.array([[1,2,4],[4,2,9]])
print(arrInt)
print(arrInt.dtype)
```

```python
arrFloat = np.array([[1,3,5],[6,4.0,9]])
print(arrFloat)
print(arrFloat.dtype)
```

```python
# itemsize -> get size of each item in the array.
arrFloat.itemsize
```

```python
arrInt.itemsize
```

```python
# To change to dtype of the array.
arr2 = np.array(["43",'4',"5"])
print(arr2)
print(arr2.dtype)
int_arr = arr2.astype(int)
print(int_arr)
print(int_arr.dtype)
```

```python
arr3 = np.array(["43.5",'4',"5"])
float_arr = arr3.astype(float)
print(float_arr)
print(float_arr.dtype)
```

```python
Str_arr = float_arr.astype(str)
print(Str_arr)
print(Str_arr.dtype)
print(Str_arr.itemsize)
```

```python
a1 = np.array([5,2,1], dtype= 'int')
print(a1)
print(a1.itemsize)
```

```python
a2 = np.array([-128,0,127], dtype= 'int8')
print(a2)
print(a2.itemsize)
```
