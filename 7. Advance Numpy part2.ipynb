### np.sort

```python
# return a sorted array of numpy array type.

import numpy as np
a = np.random.randint(1,100,20)
print(a)

np.sort(a)
```

```python
# in descending order
np.sort(a)[::-1]
```

```python
# for 2d array

b = np.random.randint(1,100,20).reshape(4,5)
print(b)

print(np.sort(b,axis=0))  # col wise sort
print(np.sort(b,axis=1))  # row wise sort
```

### np.corrcoef

### np.isin

```python
# we can check multiple item that it is in the array or not, (like membership operator)

a = np.random.randint(1,50,20)
b = np.array([2,34,9,14,31,42])

print(a)
print(b)

a[np.isin(a,b)]
```

### set Functions

- np.union1d
- np.intersect1d
- np.setdiff1d
- np.setxor1d
- np.in1d

```python
# np.union1d

a = np.array([1,2,3,4,5,6])
b = np.array([3,4,5,6,7,8])

np.union1d(a,b)
```

```python
np.intersect1d(a,b)
```

```python
np.setdiff1d(a,b)  # setdiff(a,b) -> in a but not in b
```

```python
np.setdiff1d(b,a)
```

```python
np.setxor1d(a,b)  # return (a union b) - (a intersect b)
```

```python
a[np.in1d(b,4)]
```

### np.clip

```python
# trim the values which are exceeding the limits and replace with the max and min limits

x = np.random.randint(1,100,10)
print(x)

```

```python
np.clip(x,min = 25,max= 75)
```
