```python
import numpy as np
```

```python
# Reshaping the array 
# 1D array to n dimension array.
arr = np.array([1,2,3,4,5,6])
print(arr)
print(arr.reshape(1,6))
print(arr.reshape(6,1))
print(arr.reshape(2,3))
print(arr.reshape(3,2))
```

```python
# n dimension to 1D array.
'''
2 methods.
1. ravyel() - onl view
2. flatten() - create copy
'''
arr1 = np.array([[1,2,3],[4,5,6]])
print(arr1)
print(arr1.ravel())
print(arr1.flatten())
```
