### Broadcasting

```python
# when two or more arrays are performing operations , but their sizes are not same, but somehow they work perfectly.
```

```python
## Rules for Broadcasting
```

```python
# Rule 1. -> Make the two arrays have same no. of dimensions.
# if 2 arrays are differnet dimensions then make them in same dim.
# HOW->  take the small dim array and, add 1 in front of the dim.
```

```python
# Rule 2.  -> Make each dimension of the same size.
# Note: apply rule no. 2 when any of them have 1 in the size
# take the 1 and match with the corresponding dim. 
```

```python
# CHeck ->
# Now, check if both are in same dim, then Ok
# Otherwise , error.
```

```python
# same shape
a = np.arange(6).reshape(2,3)
b = np.arange(6,12).reshape(2,3)

print(a,"\n")
print(b,"\n")

print(a+b)
```

```python
# diff shape
a = np.arange(6).reshape(2,3)
b = np.arange(3).reshape(1,3)  

# Rule 1-> check for same dim. -> Ok
# Rule 2-> check for any 1 in the size, YES, so 1 -> 2(match with the another array size.)
# Check Now, BOth are in same dim. - OK.

print(a,"\n")
print(b,'\n')

print(a+b)
```

```python
# More examples

a = np.arange(12).reshape(4,3)
b = np.arange(3)

# R1 -> Check both are in same dim- NO, so change them.
# (3,) - 1d and (4,3) - 2d , so add 1 into the 1d to make it 2d. -> (1,3)
# R2 -> check for any 1 in any of them - YES . so match that 1 with the corresponding size.
# (1,3) -> (4,3) {1->4]
# Now check same -> YES

print(a,'\n')
print(b,'\n')

print(a+b)
```

```python
a = np.arange(12).reshape(3,4)
b = np.arange(3)

# R1 -> Check both are in same dim- NO, so change them.
# (3,) - 1d and (4,3) - 2d , so add 1 into the 1d to make it 2d. -> (1,3)
# R2 -> check for any 1 in any of them - YES . so match that 1 with the corresponding size.
# (1,3) -> (3,3) {1->3]
# Now check same -> NO

print(a)
print(b)

print(a+b)
```

```python
a = np.arange(3).reshape(1,3)
b = np.arange(3).reshape(3,1)

print(a)
print(b)

print(a+b)
```

```python
a = np.arange(3).reshape(1,3)
b = np.arange(4).reshape(4,1)

print(a)
print(b)

print(a + b)
```

```python
a = np.array([1])
# shape -> (1,1)
b = np.arange(4).reshape(2,2)
# shape -> (2,2)

print(a)
print(b)

print(a+b)
```

```python
a = np.arange(12).reshape(3,4)
b = np.arange(12).reshape(4,3)

# R1 -> Check for same dim - YES
# R2 -> check for any 1 in any of the array - NO 
# so can't match size, (only, who have 1 they can change.)

print(a)
print(b)

print(a+b)
```

```python
a = np.arange(16).reshape(4,4)
b = np.arange(4).reshape(2,2)

print(a)
print(b)

print(a+b)
```

### Working with Maths Formulas

```python
a = np.arange(10)
print(a)
print(np.sin(a))
```

```python
# In numpy we have many build in maths functions. 
# But some functions are not present, so we can make it easily.

# Sigmoid
# sigmoid(x) = 1/(1+e^(-x))
def sigmoid(x):
    return 1/(1+np.exp(-(x)))

a = np.arange(100)
sigmoid(a)
```

```python
# mean square error.

y_actual = np.random.randint(1,100,25)
y_predicted = np.random.randint(1,100,25)

print(y_actual)
print(y_predicted)

# creating mse function
# mse = summation((y_act - y_pred)^2)/N
def mse(y_act,y_pred):
    return np.mean((y_act - y_pred)**2)

print("MSE : ",mse(y_actual,y_predicted))
```
