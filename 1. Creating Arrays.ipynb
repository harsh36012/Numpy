```python
import numpy as np
```

```python
# Creating numpy array

# Using np.array() // 1D,2D
# Using np.zeros()/ ones() / empty() / random()
# Using np.arange()
# Using np.linspace()
# Using copy()
# Using identity()
```

```python
# using np.array()
arr1 = np.array([1,2,4,5])
print(arr1)
arr2 = np.array([[1,2,4],[4,5,6]])
print(arr2)
```

```python
# using np.zeros()
zero = np.zeros(5)
print(zero)
```

```python
zero_2_3 = np.zeros((2,3))
print(zero_2_3)

```

```python
# using np.ones()
one = np.ones((2,3))
print(one)
```

```python
fulledArray = np.full((5,5),5)
print(fulledArray)
```

```python
# to create array of sequence of numbers. with np.arange()
arr1 = np.arange(1,10,2)
print(arr1)
```

```python
arr2 = np.arange(5)
print(arr2)
```

```python
arr3 = np.arange(0,11)
print(arr3)
```

```python
# creating identity Matrix,  with np.eye(shape)
identity_matrix = np.eye(5)
print(identity_matrix)
```

```python
# identity matrix using np.identity()
arr5 = np.identity(5)
arr5
```

```python
# # Creating diagonal Matrix. with diag()
# Diagonal_matrix = np.diag([5,2,10])
arr6 = np.diag([1,2,3])
arr6
```

```python
# using np.linspace()
arr7 = np.linspace(1,10,5)
arr7
```

```python
# using np.copy()
arr8 = np.copy(arr7)
arr8
```

```python
# using np.random
arr9 = np.random.rand(5)
arr9
```

```python
arr10 = np.random.rand(2,5)
arr10
```
